# Train IRIS on Atari Pong

IRIS is designed for sample-efficient Atari-style control with a discrete autoencoder and autoregressive world model. This tutorial uses `ALE/Pong-v5` and very small epoch counts for a smoke run.


```python
import torch
from world_models.configs.iris_config import IRISConfig
from world_models.training.train_iris import IRISTrainer

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
```

```python
cfg = IRISConfig()
cfg.total_epochs = 2
cfg.collection_epochs = 1
cfg.env_steps_per_epoch = 200
cfg.training_steps_per_epoch = 10
cfg.transformer_batch_size = 4
cfg.checkpoint_interval = 1

trainer = IRISTrainer(game="ALE/Pong-v5", device=device, seed=0, config=cfg)
trainer
```
