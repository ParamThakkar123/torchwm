# Train JEPA on CIFAR-10 images

JEPA is a self-supervised representation learner, so this notebook uses an image dataset instead of an online RL environment. CIFAR-10 is a small tutorial dataset; ImageNet-style folders are better for serious representation learning.


```python
import torch
import torchwm

print("Device:", "cuda" if torch.cuda.is_available() else "cpu")
```

```python
cfg = torchwm.create_config(
    "jepa",
    dataset="cifar10",
    batch_size=64,
    epochs=1,
    lr=1e-4,
    root_path="data",
    crop_size=32,
    download=True,
)

agent = torchwm.create_model("jepa", cfg)
agent
```
